import data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import GridSearchCV
from sklearn import svm
```

we will load the data here

```python
data = pd.read_csv('spam.csv')
```

```python
data.info()
```

```python
data.head()
```

we will know split the data into and and x and y

```python
X = data['EmailText']
Y = data['Label']
```

```python
print(X)
```

know we will split the data into training and testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_t, Y_train, Y_t = train_test_split(X,Y,test_size=0.4, random_state=42)
X_test , X_cv , Y_test , Y_cv = train_test_split(X_t,Y_t,test_size=0.5, random_state=12)
```

know we will convert the text into integer witht the help of CountVectorizer


```python
cv = CountVectorizer()
X_train = cv.fit_transform(X_train)
X_test= cv.transform(X_test)
X_cv = cv.transform(X_cv)
```

know at this point we will apply the svm algorithm 

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 32)
classifier.fit(X_train,Y_train)
```

```python
print(classifier.score(X_test,Y_test))
```

```python
print(classifier.score(X_train,Y_train))
```

```python
print(classifier.score(X_cv,Y_cv))
```
